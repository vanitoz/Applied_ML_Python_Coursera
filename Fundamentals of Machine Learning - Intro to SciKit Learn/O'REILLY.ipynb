```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (iris_dataset['data'],iris_dataset['target'], random_state=0)
```

```python
# random_state=0 - pseudorandom numb. generator - to make sure we will get same output if we run func. several times
```

```python
import pandas as pd
from matplotlib import cm

iris_df=pd.DataFrame(X_train, columns=iris_dataset.feature_names)
```

```python
cmap = cm.get_cmap('gnuplot')
pd.plotting.scatter_matrix(iris_df, c=y_train, figsize=(15,15), marker='o', hist_kwds={'bins':20}, s=60, alpha=.8, cmap=cmap)
```
