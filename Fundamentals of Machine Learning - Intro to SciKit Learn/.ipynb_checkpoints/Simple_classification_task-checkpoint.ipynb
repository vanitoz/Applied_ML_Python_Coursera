```python
#%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split

fruits = pd.read_table('fruit_data_with_colors.txt')
```

```python
# Spliting original data in 2 parts (Creating Training and Testing Sets(to evaluate trained classifier))
# Train test split function randomly shuffels date and splits off a certain percentage of the input samples.
# 75-25% split most common (standart)
```

```python
X = fruits [['mass','width','height','color_score']]  # holds the features of our data set without the lable
y = fruits ['fruit_label']

X_train, X_test, y_train, y_test = train_test_split (X, y, random_state=0) # random_st - provides a seed value to the function's internal
                                                                            # rendom number generator. randomize splits 
```

```python
# k-Nearest eighbors algorithm (k-NN) can be used for clasification and regression!!!
# k-NN - instance based or memory based supervised learning (memorising labeled examples from the training set)
# and then they use those memorized exammples to classify new object later
# k reffers to the number of nearest neighbors the classifier will retrive and use in order to make its prediction

```

```python
# For this example, we use the mass, width, and height features of each fruit instance
X = fruits[['mass', 'width', 'height']]  # training set (calls feature space)
y = fruits['fruit_label'] # training lables set

# default is 75% / 25% train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Create classifier object 

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
# Training classifier (fit the estimator) using our training data
# k-nn classifier in this case is an example of general class - called an estimator 
# all estimators has a method fit. that take training data and changes the state of classifier
knn.fit(X_train,y_train)  # using fit method. classifier will memorize the training set examples in kinda internal storage
```

```python
# no as we have already trained classifier we can check how accurate it's likely to be on some new instances
# to do that we can apply our clasifier to all instances in our test sets that not included in classifier training
```

```python
knn.score(X_test,y_test)  #  compute the accuracy (fraction of test set items whose true label was correctly predicted)
```

```python
# use trained k-NN classifier model to classify new, previously unseen objects
```

```python
fruit_prediction=knn.predict([[20,4.3,5.5]])
lookup_fruit_name[fruit_prediction[0]]
```

```python
fruit_prediction=knn.predict([[100,6.3,8.5]])
lookup_fruit_name[fruit_prediction[0]]
```

```python
k_range = range(1,20)
scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    scores.append(knn.score(X_test, y_test))

plt.figure()
plt.xlabel('k')
plt.ylabel('accuracy')
plt.scatter(k_range, scores)
plt.xticks([0,5,10,15,20]);
```

```python
t = [0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2]

knn = KNeighborsClassifier(n_neighbors = 5)

plt.figure()

for s in t:

    scores = []
    for i in range(1,1000):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1-s)
        knn.fit(X_train, y_train)
        scores.append(knn.score(X_test, y_test))
    plt.plot(s, np.mean(scores), 'bo')

plt.xlabel('Training set proportion (%)')
plt.ylabel('accuracy');
```
